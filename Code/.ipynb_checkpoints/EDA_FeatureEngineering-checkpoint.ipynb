# Feature Selection

### Categorical Variables 
We will first try to find the association between our categorical variables and Target Variable(Approved Flag)


```python
cat_var = []
for i in df.columns:
    if df[i].dtype == 'object' and i != 'Approved_Flag':
        cat_var.append(i)
cat_var
```

**Hypothesis Testing on Catigorical Featuures**   
- `Chi-Square` is the Test which we use to find association between Categorical - Categorical Variable.   
- We will assume `alpha` = 0.05

```python
# let's calculate p-value using chi-square test
```

```python
# Chi-Square Test
p_val = []
for i in cat_var:
    chi2, pval, _, _, = chi2_contingency(pd.crosstab(df[i], df['Approved_Flag']))
    print(i, ':', pval)
    p_val.append(pval)
```

```python
for i in p_val:
    print(p_val[0] < 0.05)
```

**As we can see calculated p-value sor all the categorical values is less than assumed alpha, we will accept all.**

### Numerical Variables 

Now we will try to find association between numerical variables and target variable with the use of appropriate Statistical Test

```python
num_var = []

for i in df.columns:
    if df[i].dtype != 'object' and i not in ['PROSPECTID', 'Approved_Flag']:
        num_var.append(i)
num_var
```

**Multicollinearity check in Numerical features**
- We need to check if our independent Variables possess multicollinearity, if we can predict any independent feature using other independent variables, then we can not use that feature because it will wrongly interpret our model
- For ex: If we are predicting a house price, and we take two independent variable one is `Floor Area` other is `Carpet Area`, we can predict carpet area from floor area, if we use both in house price prediction it will be a misleading interpretation to the model.

**To check multicollinearity we will regress each variable with others**
- We will use `VIF(Variance Inflation Factor)` to identify multicollinearity among independent variables.
- It Takes R-Squared values for each Independent variable and eliminate if crosses the threshold.  
- `VIF = 1/(1 - r^2)`

```python
# VIF sequentially check

vif_data = df[num_var]
total_columns = vif_data.shape[1]
columns_to_be_kept = []
column_index = 0



for i in range (0,total_columns):
    
    vif_value = variance_inflation_factor(vif_data, column_index)
    print (column_index,'---',vif_value)
    
    
    if vif_value <= 6:
        columns_to_be_kept.append( num_var[i] )
        column_index = column_index+1
    
    else:
        vif_data = vif_data.drop([ num_var[i] ] , axis=1)
```

```python
len(columns_to_be_kept)
```

**`Observation`**
- We have used `Sequential` method to check VIF values 
- From `72` numerical collumns, we are left with `39` columns, rest of the columns have `VIF` value more than `6`, Representing `High` Multicollinearity


**`Note`**
1. **Parallel**
    - In Parallel method, we calculate VIF for each independent variable(say it target) with every other variable. 
    - In the end if we have three features having high multicollinearity, we will remove all three.   
    ~**`Limitation`**
    - If we found 3 variables with high multicollinearity with each other, we do not need to remove all three variables, if we remove two then third column won't have any column to generate multicollinearity with.
    - In this method, there is high chance we will remove extra columns more than needed.
2. **Sequential**
    - In this method if we find any variable with high collinearity with any other variable, we will remove it right away, in this way it will not remove every feature with high multicollinearity(which is not required)
    
    
More Relevant features we have in our data, more accurate will be our results

**Hypothesis Testing on Numerical Featuures**  
- We will use Anova test to calculate p-value and then compare the p-value of each column with `alpha`(0.05).  


**Reason To use Anova test**
- To find association between numerical and categorical variable, `Anova` and `T-test` are most commonly used tests.
- As we know our target variable is a categorical variable, with 4 categories `p1`, `p2`, `p3`, `p4`.   

1. `Anove`: When categorical variable has 3 or more than 3 categories.   
2. `T-test`: When categorical variable has 2 categories.

```python
columns_to_be_kept_num = []
for i in columns_to_be_kept:
    a = list(df[i])
    b = list(df['Approved_Flag'])
    
    group_p1 = [value for value, group in zip(a, b) if group == 'P1']
    group_p2 = [value for value, group in zip(a, b) if group == 'P2']
    group_p3 = [value for value, group in zip(a, b) if group == 'P3']
    group_p4 = [value for value, group in zip(a, b) if group == 'P4']
    
    f_stats, p_value = f_oneway(group_p1, group_p2, group_p3, group_p4)
    
    if p_value <= 0.05:
        columns_to_be_kept_num.append(i)
        
    
```

```python
len(columns_to_be_kept_num)
```


**After applying Anova test on 39 numerical variables, we are left with 37 variables.**

```python
final_features = columns_to_be_kept_num + cat_var
df = df[final_features + ['Approved_Flag']]
df.shape
```
